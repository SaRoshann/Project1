```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_dataset(file_path):

  return pd.read_csv(file_path)
  df = load_dataset("D:/AI.ML/python/titanic.csv")
df.head(10)
```

## Line plot

```python
sns.lineplot(x='age', y='SysBP', data=df, color='red', marker='.', linestyle=':')
plt.title('Relationship Age Height and Systolic Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Systolic Blood Pressure')

plt.show()
```

#### Systolic Blood Pressure means the pressure at the arteries of our heart, where it is pumping the blood out into the body at maximum pressure.On the x-axis we have Age, comparing these two we can tell when the patient crosses high blood pressure. We can see that the age group 60 has high Bp putting them at higher risk, also if we see a spike in age group of 20-30 which can be called outlier, it tells doctors that Bp has increased due to unnatural(not aging) agent like stress.

## Histogram

```python
plt.hist(df['BMI'] ,color='pink')
```

#### Above is the representation of histogram, we can see that the histogram is left-skewed which means the number of people having healthy BMI(18-24) is less , which hence indicates that more number of people have unhealthy BMI.

## Barplot

```python
sns.barplot(x='sex',y='DiaBP', color='pink',data=df)
```

#### Above is the representation of Barplot between the sex (0=male,1=female). We can clearly see that the average DiaBP for both is same, which may indicate that sex doesnt affect one's DiaBp.

## Boxplot

```python
sns.boxplot(df['HR'], color='pink')
```

#### Above is the representation of Box plot for the Heart rate category. We can clearly see that the average Heart rate is 140, which shows the data is taken after a physical activity as normal resting HR is (60-100).Also if we observe there are no outliers which cleary shows there cannot be a outlier in this case as one cannot have a unnormal heart rate

## PieChart

```python
df['sex'].value_counts().plot(
    kind='pie', 
    autopct='%1.1f%%', 
    colors=['lightblue', 'pink']
)
```

#### Above is the representation of pie-chart where the chart represents the percentage of males and females(0=male,1=female), who appeared for the health checkup.

## ScatterPlot

```python
sns.scatterplot(x='sex', y='weightKg', data=df, color='pink')
```

#### Above is the representation of scatter plot, why i took sex and weight is beacuse if i take any other parameter than sex the data is too clustered and difficult to read, here we can see the weight distrucution of both males and females
